```python
from bs4 import BeautifulSoup
import requests

url = "https://www.iranjib.ir"
response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')
headline_bs4 = soup.find('div', id = "column2")
h3_soup = headline_bs4.find_all('h3')
news_urls = []
for each_h3 in h3_soup:
    news_urls.append(each_h3.find('a').get('href'))
```

```python
h3_soup[0].find('a').get('href')
```

```python
len(news_urls), len(set(news_urls))
```

```python
with open('iranjib_data.txt', 'w', encoding='utf-8') as file:
    file.write('id, title, summary, content, date, views\n')
    for id, page_url in enumerate(news_urls):
        page_response = requests.get(page_url)
        page_soup = BeautifulSoup(page_response.text, 'html.parser')
        title = page_soup.find('h1').get_text()
        summary = page_soup.find('div',{'class':'newssummary'}).get_text()
        matn_soup = page_soup.find('div', {'class':'matn'})
        paragraphs = matn_soup.find_all('p')
        news_text = ''
        for p in paragraphs:
            news_text += p.get_text() + '\n'
        table_soup = page_soup.find_all('table')[1]
        tr_soup = table_soup.find_all('tr')[1]
        date = tr_soup.find_all('td')[1].get_text()
        views = tr_soup.find_all('td')[2].get_text()
        file.write(f'{id}, {title}, {summary}, {news_text}, {date}, {views}\n')
        
```
